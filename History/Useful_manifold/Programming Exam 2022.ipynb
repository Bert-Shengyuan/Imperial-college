```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from matplotlib.widgets import Slider

# Create some sample data
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Create a figure and 3D axis
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot initial data
scatter = ax.scatter(x, y, z, c=z, cmap='viridis')

# Add sliders for interactivity
axcolor = 'lightgoldenrodyellow'
ax_x = plt.axes([0.25, 0.02, 0.65, 0.03], facecolor=axcolor)
ax_y = plt.axes([0.25, 0.07, 0.65, 0.03], facecolor=axcolor)

slider_x = Slider(ax_x, 'X-axis', min(t), max(t), valinit=min(t))
slider_y = Slider(ax_y, 'Y-axis', min(t), max(t), valinit=min(t))

# Update function for animation
def update(frame):
    ax.view_init(elev=20, azim=frame)
    return scatter,

# Update sliders
def update_sliders(val):
    ax.view_init(elev=20, azim=val)
    fig.canvas.draw_idle()

slider_x.on_changed(update_sliders)
slider_y.on_changed(update_sliders)

# Create animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 360, 1), interval=50, blit=True)

plt.show()
```
